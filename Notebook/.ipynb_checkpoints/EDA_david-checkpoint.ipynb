feature engineering: 
1. value count
2. stat encoding / time window (mean, median, max, min, std)
3. diff of mean with adjeacent window
4. last appearance 
5. categocial key
6. card_num
7. days between transaction
8. count for csmcu/stocn to identify dominant type
9. floating points

feature name transform:
1. acqic $\rightarrow$ acquirer
2. bacno $\rightarrow$ bank_num
3. cano $\rightarrow$ card_num
4. contp $\rightarrow$ trans_category
4. csmcu $\rightarrow$ coin_type
5. ecfg $\rightarrow$ online
6. etymd $\rightarrow$ trans_type
7. flbmk $\rightarrow$ fallback
8. flg_3dsmk $\rightarrow$ 3ds
9. hcefg $\rightarrow$ pay_type
10. insfg $\rightarrow$ installment
11. mcc
12. mchno $\rightarrow$ mch_num
13. ovrlt $\rightarrow$ excess
14. scity $\rightarrow$ city
15. stocn $\rightarrow$ nation
16. stscd $\rightarrow$ status
17. txkey $\rightarrow$ key

# Covariate Shift

```python
# check the distribution difference between test and train
not_calc = ['absolute_time', 'minute', 'hour', 'second', 'locdt', 'loctm', 'fraud_ind', 'txkey']
for col in combine.columns:
    if(col not in not_calc):
        covariate_shift(combine, col, TRAIN_SHAPE)
```

```python
for col in combine.columns:
    if(col not in not_calc):
        print('*' * 20)
        print(col)
        train_unique = combine.loc[:TRAIN_SHAPE, col].unique()
        test_unique = combine.loc[:TRAIN_SHAPE, col].unique()
        print(train_unique.shape, test_unique.shape)
        print('Train unique:' , [x for x in train_unique if x not in test_unique])
        print('Test unique:' , [x for x in test_unique if x not in train_unique])
```
